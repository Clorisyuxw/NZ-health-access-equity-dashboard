# Power BI Dashboard Design

This notebook documents the planned Power BI dashboard structure for the Health Access and Equity Reporting Dashboard.

The dashboard uses validated PostgreSQL reporting views exported as CSV files.

The purpose of the dashboard is to communicate:
- which health access barriers are most common
- which barriers affect the largest estimated populations
- where equity gaps are largest
- which barriers are increasing or decreasing over time
- which issues should be prioritised for service planning

## Dashboard Structure

The Power BI dashboard is designed with four pages:

1. Access Barrier Overview
2. Priority Matrix
3. Equity Gap Analysis
4. Trend Over Time

Each page answers a different analytical question.

## Page 1: Access Barrier Overview

### Analytical Question

Which health access barriers were most common in the latest reporting year?

### Main Dataset

- `vw_latest_prevalence.csv`
- `vw_priority_barriers.csv`

### Recommended Visuals

1. Bar chart: prevalence percentage by indicator
2. Bar chart: estimated affected population by indicator
3. Cards: top adult barrier and top child barrier
4. Slicers: population and barrier type

### Key Fields

- `population`
- `indicator_short_name`
- `barrier_type`
- `total_value`
- `estimated_number`
- `prevalence_rank`

## Page 2: Priority Matrix

### Analytical Question

Which health access barriers should be prioritised based on prevalence, equity gap and trend?

### Main Dataset

- `vw_priority_matrix.csv`

### Recommended Visuals

1. Matrix table: indicator by priority category
2. Scatter plot: prevalence value vs maximum adjusted rate ratio
3. Bar chart: average annual percentage point change by indicator
4. Slicers: population, barrier type and priority category

### Key Fields

- `population`
- `indicator_short_name`
- `total_value`
- `max_adjusted_rate_ratio`
- `average_annual_pp_change`
- `priority_category`
- `prevalence_level`
- `equity_gap_level`
- `trend_level`

### Important Note

The priority categories are analyst-defined and are used for exploratory reporting and service planning. They are not official Ministry of Health classifications.

## Page 3: Equity Gap Analysis

### Analytical Question

Which population groups experience higher health access barriers?

### Main Dataset

- `vw_equity_gap.csv`

### Recommended Visuals

1. Bar chart: adjusted rate ratio by comparison
2. Table: comparison, indicator, adjusted rate ratio and adjusted confidence interval
3. Slicers: population, barrier type and indicator

### Key Fields

- `population`
- `indicator_short_name`
- `barrier_type`
- `comparison`
- `adjusted_rate_ratio`
- `adjusted_rate_ratio_low_ci`
- `adjusted_rate_ratio_high_ci`
- `adjusted_for`

### Interpretation Note

Adjusted rate ratios compare one population group with a reference group after adjustment.

A value greater than 1 suggests the comparison group had a higher rate of the access barrier than the reference group.

## Page 4: Trend Over Time

### Analytical Question

Are selected health access barriers increasing or decreasing over time?

### Main Datasets

- `fact_trend.csv`
- `vw_trend_summary.csv`

### Recommended Visuals

1. Line chart: percentage value by year
2. Bar chart: average annual percentage point change by indicator
3. Table: start year, end year, start value, end value and percentage point change
4. Slicers: population, barrier type and indicator

### Key Fields

From `fact_trend.csv`:
- `year`
- `percent_value`
- `population`
- `group`
- `indicator_short_name`
- `barrier_type`

From `vw_trend_summary.csv`:
- `start_year`
- `end_year`
- `years_observed`
- `percentage_point_change`
- `average_annual_pp_change`
- `trend_direction`

### Interpretation Note

Trend comparisons are based on available years, and different indicators may have different starting years.
